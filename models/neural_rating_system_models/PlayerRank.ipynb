```python
new_pos = []
for i, row in df[['Pos']].iterrows():
    pos = row['Pos']
    counts = {
        'D': 0,
        'M': 0,
        'A': 0, 
        'G': 0
    }
    for p in pos.split(','):
        counts[map_dict[p]] += 1
    max_val = 0
    best_key = 'M'
    for key in counts:
        if counts[key] > max_val:
            max_val = counts[key]
            best_key = key
    new_pos.append(best_key)
    
#     print(pos, best_key)
```

```python
per_match_ratings = df.groupby(['Season', 'League', 'Team', 'Date', 'Pos']).sum()[offensive_columns]
per_match_ratings['rating'] = clf['standardscaler'].transform(per_match_ratings).dot(clf['svc'].coef_.T).T[0]
per_match_ratings = per_match_ratings.reset_index()
per_match_ratings
```

```python
final_ratings = []
for key, data in per_match_ratings.groupby(['Season', 'League', 'Team', 'Date']):
    current_row = {
        'Season': key[0],
        'League': key[1],
        'Team': key[2],
        'Date': key[3],
        'ARating': 0,
        'DRating': 0,
        'MRating': 0,
        'GRating': 0
    }
#     print(data['Pos'])
#     print(data['rating'])
    for i in range(len(data)):
        current_row[data['Pos'].iloc[i]+'Rating'] = data['rating'].iloc[i]
    final_ratings.append(current_row)
```

```python
final_ratings_df = pd.DataFrame(final_ratings)
```

```python
final_ratings_df = pd.merge(
    final_ratings_df, 
    matches_df_long, 
    how='inner', 
    on=['League', 'Season', 'Team', 'Date']
).drop(['Match Report', 'Referee', 'Win'], 1)
```

```python
from sklearn.metrics import brier_score_loss

import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
# Confirming with the results shown in class
cols_for_pred = [
    'ARatingDiff', 
    'DRatingDiff',
    'MRatingDiff',
    'GRatingDiff'
]

glm_str = "Win ~ " + '+'.join(cols_for_pred)

result_glm = smf.glm(glm_str, 
             data = final_matches_df[final_matches_df['Season'] == '2017-2018'], 
             family = sm.families.Binomial()
).fit()
result_glm.summary()
```

```python
print("Test Brier Score: ", brier_score_loss(final_matches_df[final_matches_df['Season'] == '2018-2019']['Win'], result_glm.predict(final_matches_df[final_matches_df['Season'] == '2018-2019'])))
```
