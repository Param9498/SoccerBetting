```python
import pandas as pd
```

```python
fbref_df = pd.read_csv('fbref/merged_ODrating_fbref.csv', parse_dates=['Date']).sort_values('Date')
understat_df = pd.read_csv('understat/matches.csv', parse_dates=['Date']).sort_values('Date')
_538_df = pd.read_csv('538/matches.csv', parse_dates=['Date']).sort_values('Date')
fbdata_df = pd.read_csv('fbdata/matches.csv', parse_dates=['Date']).sort_values('Date')
```

```python
merge1 = pd.merge(fbref_df, understat_df, on = ['League', 'Season', 'Team_Home', 'Team_Away', 'Date'], how = 'inner')
merge1
```

```python
merge2 = pd.merge(merge1, _538_df, on = ['League', 'Season', 'Team_Home', 'Team_Away', 'Date'], how = 'inner')
merge2
```

```python
merge3 = pd.merge(merge2, fbdata_df, on = ['League', 'Season', 'Team_Home', 'Team_Away', 'Date'], how = 'inner')
merge3
```

```python
merge3 = merge3.sort_values('Date').reset_index(drop=True)
```

```python
merge3 = merge3.drop(['Unnamed: 0', 'Goals_Home_x', 'Goals_Away_x'], 1)
merge3 = merge3.rename(columns = {'FBREFxg_away': 'FBREFxg_Away', 'Goals_Home_y': 'Goals_Home', 'Goals_Away_y': 'Goals_Away'})
```

```python
merge3.to_csv('final_data.csv', index = False)
```
